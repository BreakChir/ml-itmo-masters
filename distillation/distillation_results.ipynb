```python
import matplotlib.pyplot as plt
```

```python
def parse_float(line):
    return float(line.strip())
```

```python
def read_accs(model_name):
    file_path = 'logs/{0}.log'.format(model_name)
    with open(file_path, 'r') as f:
        lines = f.readlines()
        return list(map(parse_float, lines))
```

```python
student_accs = read_accs('student') # 2 layers, with distillation, alpha = 0.5
teacher_accs = read_accs('teacher') # 5 layers
simple_accs = read_accs('simple') # 2 layers, without distillation
```

```python
def get_range(accs):
    n = len(accs)
    return list(range(n))
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(get_range(student_accs), student_accs, label='student accuracy')
plt.plot(get_range(teacher_accs), teacher_accs, label='teacher accuracy')
plt.plot(get_range(simple_accs), simple_accs, label='no distillation accuracy')
plt.legend(loc=4, prop={'size': 15})
plt.show()
```
