```python
import matplotlib.pyplot as plt
```

```python
def parse_float(line):
    return float(line.strip())
```

```python
def read_accs(file_name):
    file_path = f'../logs/swarm-sgd/{file_name}.txt'
    with open(file_path, 'r') as f:
        lines = f.readlines()
        return list(map(parse_float, lines))
```

```python
exp_name = [f'k={k}' for k in [1, 5, 10, 20]]
```

```python
exps = {name: read_accs(name) for name in exp_name}
```

```python
def get_range(accs):
    n = len(accs)
    return list(range(n))
```

```python
def show_results(exps, exp_names):
    plt.rcParams["figure.figsize"] = (20,10)
    for exp_name in exp_names:
        cur_accs = exps[exp_name]
        plt.plot(get_range(cur_accs), cur_accs, label=exp_name)
    plt.legend(loc=4, prop={'size': 15})
    plt.show()
```
